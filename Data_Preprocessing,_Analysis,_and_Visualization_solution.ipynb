# **Data Preprocessing**

First we have to remove the unnecessary attributs like: CustomerId, Surname, RowNumber, because we don't need these informations in learning phase.
To do that we use the function drop() by giving it as inputs the list of the attributs to be deleted and the axis, which is 1 in this case.


```python
#write the code that drop the attributs CustomerId, Surname, RowNumber from the data 
data = data.drop(['CustomerId', 'Surname', 'RowNumber'], axis = 1)
print(data.columns)
```

To train the a machine learning model we have to split the data into X(independent variables, or the features) and Y(dependent variables, or the labels), and we can do that using the iloc function

.iloc â†’ means index location, i.e. select by row/column number, not by name.

: before the comma â†’ all rows

0:10 after the comma â†’ columns 0 through 9 (Python slicing stops before 10)

So x will contain:

all rows, and the first 10 columns of data.

```python
#Split the data into X and Y here
x = data.iloc[:,0:10]
y = data.iloc[:,10]

print(x.shape)
print(y.shape)

print(x.columns)
```

Also, to make the learning process smoother we need to change Categorical variables into numerical variables using pd.get_dummies() function

```python
#Turn categorical variables into numerical variables here
x = pd.get_dummies(x)

x.head()
```

Now, we need to split the data into training and test set with portion of 0.25 for test set and 0.75 for training set. To do that we have to use the function **train_test_split** from the package sklearn.model_selection

```python
#Split the data here
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```
